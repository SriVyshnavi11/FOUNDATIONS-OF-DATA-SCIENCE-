```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/Automobile_data.csv")

```

```python
print("First 5 Records")
print(df.head())
```

```python
print("\n Dataset Shape")
print(df.shape)
```

```python
print("\nColumn Names")
print(df.columns)
```

```python
print("\nData Types")
print(df.dtypes)
```

```python

print("\nMissing Values")
print(df.isnull().sum())
```

```python
print("\nStatistical Summary")
print(df.describe())
```

```python
sns.countplot(x='fuel-type', data=df)
plt.title("Fuel Type Distribution")
plt.show()
```

```python
plt.figure(figsize=(12,5))
sns.countplot(x='make', data=df)
plt.xticks(rotation=90)
plt.title("Car Company Distribution")
plt.show()
```

```python
sns.countplot(x='body-style', data=df)
plt.title("Body Style Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['horsepower'], bins=20, kde=True)

plt.title("Horsepower Distribution")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['price'], bins=20, kde=True)

plt.title("Price Distribution")
plt.show()
```

```python
plt.figure(figsize=(10,8))

sns.heatmap(df.corr(numeric_only=True),
            annot=True,
            cmap='coolwarm')

plt.title("Correlation Matrix")
plt.show()
```

```python
print("Numeric Attributes")

numeric = df.select_dtypes(include=['int64','float64']).columns
print(numeric)

print("\nNominal Attributes")

nominal = ['fuel-type','body-style']
print(nominal)

print("\nBinary Attributes")

binary = ['engine-location']
print(binary)
```
